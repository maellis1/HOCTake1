```python
import json
import requests
import pandas
```

```python
url = 'https://csgenome.org/api/benchmarks/top500?page=1&limit=25000&columns=computer'
data = requests.get(url)
thisDict = data.json()
```

Next, we're going to take what we learned about conditional logic and apply it to our data set. We start by making a new dictionary to count the number of times each supercomputer makes the top 500 list. Then we'll use a loop to run through each computer in the list. If the computer already has a counter, we increment it by one. If it is a new computer, we start its counter at one. 

```python
counts = dict()
for i in range(len(thisDict["data"])):
    val = thisDict["data"][i]["computer"]
    if val in counts:
        counts[val] = counts[val] + 1
    else:
        counts[val] = 1
```

The "sorted" function is built-in in Python. It it extremely helpful for internally sorting lists, tuples, and dictionaries. In this instance, we are using this sorted function to sort the keys (the computer names) by their values (the counts) in a descending order. This will let us show the most common Top 500 supercomputers.

```python
sortedCounts = sorted(((value, key) for (key,value) in counts.items()), reverse=True)
```

Now we're going to make the dataframe that will display our data. The index list starts our counter at 1, instead of the default counter which starts at 0. For a refresher on how pandas dataframes work, use this link (https://www.datacamp.com/community/tutorials/pandas-tutorial-dataframe-python)and do other research to make sure you understand the rest of the code.

```python
idx = []
for x in range(0, len(sortedCounts)):
    idx.append(str(x+1) + ".")
sortedDF = pandas.DataFrame(sortedCounts, index=idx, columns=['Count', 'Computer'])
sortedDF = sortedDF[['Computer', 'Count']]
sortedDF.head(50)
```
